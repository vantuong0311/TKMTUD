Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

```python
import numpy as np

# Phát sinh mảng ndarray ngẫu nhiên gồm 100 phần tử, giá trị từ 0 đến 1000
arr = np.random.randint(0, 1001, size=100)

# Hiển thị 10 phần tử đầu tiên
print("10 phần tử đầu tiên:")
print(arr[:10])

# Hiển thị 5 phần tử cuối cùng
print("\n5 phần tử cuối cùng:")
print(arr[-5:])

# Lọc các phần tử chia hết cho 5 và lớn hơn 100
filtered_elements = arr[(arr % 5 == 0) & (arr > 100)]
print("\nCác phần tử chia hết cho 5 và lớn hơn 100:")
print(filtered_elements)
```

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

```python
import numpy as np

# Phát sinh mảng ndarray ngẫu nhiên gồm 100 phần tử, giá trị từ 0 đến 1000
arr = np.random.randint(0, 1001, size=100)

# Lọc các phần tử có giá trị nhỏ hơn 100
elements_less_than_100 = arr[arr < 100]

# Đếm số lượng phần tử nhỏ hơn 100
count_less_than_100 = len(elements_less_than_100)

# Hiển thị kết quả
print(f"\nSố lượng phần tử nhỏ hơn 100: {count_less_than_100}")
print("Các phần tử nhỏ hơn 100:")
print(elements_less_than_100)
```

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

```python
import numpy as np

# Phát sinh mảng ndarray ngẫu nhiên gồm 100 phần tử, giá trị từ 0 đến 1000
arr = np.random.randint(0, 1001, size=100)

# Hiển thị mảng ban đầu
print("Mảng ban đầu:")
print(arr)

# Tính tổng mảng
total = np.sum(arr)
print(f"\nTổng mảng: {total}")

# Tính giá trị trung bình
mean = np.mean(arr)
print(f"Giá trị trung bình: {mean:.2f}")

# Tính giá trị trung vị
median = np.median(arr)
print(f"Giá trị trung vị: {median}")

# Tính Q1, Q2, Q3
Q1 = np.percentile(arr, 25)  # Phân vị thứ 25
Q2 = np.percentile(arr, 50)  # Phân vị thứ 50 (giống trung vị)
Q3 = np.percentile(arr, 75)  # Phân vị thứ 75
print(f"Q1: {Q1}")
print(f"Q2 (Trung vị): {Q2}")
print(f"Q3: {Q3}")

# Tính phân vị 20 (percentile 20)
P20 = np.percentile(arr, 20)
print(f"Phân vị 20: {P20}")
```

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

```python
import numpy as np
# Phát sinh mảng ndarray ngẫu nhiên gồm 100 phần tử, giá trị từ 0 đến 1000
arr = np.random.randint(0, 1001, size=100)

# Chọn ngẫu nhiên 10 phần tử không thay thế từ mảng 5 lần
for i in range(1, 6):  # Thực hiện 5 lần
    selected_elements = np.random.choice(arr, size=10, replace=False)
    print(f"Lần chọn thứ {i}: {selected_elements}")
```

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

```python
import numpy as np
from collections import Counter

# Phát sinh mảng ndarray ngẫu nhiên gồm 100 phần tử, giá trị từ 0 đến 1000
arr = np.random.randint(0, 1001, size=100)

# Sử dụng Counter để đếm tần suất xuất hiện các phần tử
counts = Counter(arr)

# Tìm phần tử xuất hiện nhiều nhất
most_common_element, frequency = counts.most_common(1)[0]

# Hiển thị kết quả
print(f"Phần tử xuất hiện nhiều nhất: {most_common_element}")
print(f"Tần suất xuất hiện: {frequency}")
```

### Câu 4: Định dạng đường thẳng

Phát sinh x là các điểm dữ liệu từ $[0, \pi]$. Bạn hãy vẽ đồ thị $y = sin(x)$ và $y = cos(x)$ như dưới đây.
Gợi ý: sử dụng hằng số pi và hàm sau:
- `np.pi` : số $\pi$
- `np.linspace()`
- `np.sin()`
- `np.cos()`

```python
# This cell contains output. Don't remove!
```

```python
# YOUR CODE HERE
import numpy as np
import matplotlib.pyplot as plt

# Tạo dữ liệu x từ 0 đến π
x = np.linspace(0, 6, 100)

# Tính giá trị y cho hai hàm sin(x) và cos(x)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Vẽ đồ thị của y = sin(x)
plt.plot(x, y_sin, label="sin(x)", color="blue", linestyle="-", markersize=4)

# Vẽ đồ thị của y = cos(x)
plt.plot(x, y_cos, label="cos(x)", color="red", linestyle="--", markersize=4)

plt.ylim(-1, 1.1)
plt.xlim(-0.3, 6.3)

# Thêm các thành phần vào đồ thị
plt.title("Đồ thị hàm số sin(x) và cos(x)")  # Tiêu đề
plt.axhline(0, color="black", linewidth=0.8)      # Đường trục hoành
plt.axvline(0, color="black", linewidth=0.8)      # Đường trục tung
plt.legend()                                      # Hiển thị chú thích
plt.grid(True)                                    # Thêm lưới

# Hiển thị đồ thị
plt.show()
```

```python
# YOUR CODE HERE
import numpy as np
import matplotlib.pyplot as plt

# Phát sinh 10,000 số ngẫu nhiên từ phân phối chuẩn
data = np.random.normal(size=10000)

plt.figure(figsize=(10, 7))

# Vẽ boxplot theo chiều dọc
plt.boxplot(data, patch_artist=True, boxprops=dict(facecolor='white', color='black'))

# Thêm tiêu đề và nhãn
plt.title('Boxplot', fontsize=14)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Hiển thị đồ thị
plt.show()
```

### Câu 10: Kết hợp trên cùng một đồ thị

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ hình thể hiện phân phối như sau:

Hướng dẫn: thêm tham số `vert=False` trong hàm vẽ `boxplot()`

```python
# This cell contains output. Don't remove!
```

```python
# YOUR CODE HERE
import numpy as np
import matplotlib.pyplot as plt

# Phát sinh 10,000 số ngẫu nhiên từ phân phối chuẩn
data = np.random.normal(size=10000)

# Tạo đồ thị với 2 biểu đồ con (2 hàng, 1 cột)
plt.figure(figsize=(10, 7))

# Biểu đồ 1: Histogram
plt.subplot(2, 1, 1)  # (rows, cols, index)
plt.hist(data, bins=150, edgecolor='black', color='purple', alpha=0.5)  # Điều chỉnh bins để tạo khoảng cách
plt.title('Histogram', fontsize=14)
plt.ylabel('Mật độ', fontsize=12)

# Biểu đồ 2: Boxplot
plt.subplot(2, 1, 2)  # (rows, cols, index)
plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='white', color='black'))
plt.title('Boxplot', fontsize=14)

# Hiển thị đồ thị
plt.tight_layout()  # Đảm bảo các biểu đồ không bị chồng lên nhau
plt.show()
```
